```python
import matplotlib.pyplot as plt
from math import *
```

```python
def f(x):
    return a*sin(x)


def g(x):
    return b*exp(-x/c)*cos(d*x)

a = 3
b = 4
c = 5
d = 4

x1 = [x/10 for x in range(100)]
y1 = [f(x) for x in x1]
y2 = [g(x) for x in x1]
xaxis = [0 for x in x1]
```

```python
plt.plot(x1,y1,label = 'First Data Set')
plt.plot(x1,y2,label = 'Second Data Set')
plt.plot(x1,xaxis)
plt.xlabel('Independent Var')
plt.ylabel('Dependent Var')
plt.title("My First PLOT\nIs'nt it cool")
plt.legend()
plt.show()
```
